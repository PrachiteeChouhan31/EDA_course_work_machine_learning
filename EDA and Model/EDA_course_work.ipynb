Dealing with outliers for attribute X3 bus station

```python
Q1= np.percentile(data_df['X3 bus station'],25,interpolation='midpoint')
Q3= np.percentile(data_df['X3 bus station'],75,interpolation='midpoint')
IQR=Q3-Q1
lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
```

```python
print(Q1,Q3,IQR,lower_bound,upper_bound)
```

```python
#find the outliers with z-score
from scipy import stats
z=np.abs(stats.zscore(data_df['X3 bus station']))
```

```python
print(np.where(z>3))
```

# Modelling

```python
#Modelling with attribute 'X2 house age'
```

```python
#split the data into train and test parts
train_df, test_df = train_test_split(data_df,shuffle = True, test_size = 0.95, random_state=17)


X= train_df['X2 house age']
y = train_df['Y price sq feet']
X2= np.array(X).reshape(-1,1) # reshape(-1,1) makes the array horizontal, y is vertical
    
# Train the model (we use all data for training --> 
model_X2 = sklearn.linear_model.LinearRegression()
model_X2.fit(X2, y)
r_sq2 = model_X2.score(X2, y)
print(r_sq2)
plt.scatter(X2, y, color='black', label='observed')

plt.plot(X2, model_X2.predict(X2), label='fit', color='Green', linewidth=2)

plt.xlabel('X2 house age')
plt.ylabel('Y price sq feet')
plt.title('Regression on training data')
plt.legend(loc='best')

plt.show()
```

```python
# Make a prediction
X3= test_df['X2 house age']
 
X3= np.array(X3).reshape(-1,1)
# what is the charges for element 44 of the test bmi list. 
predicted_charge = model_X2.predict([X3[44]])
print('house age= {:.2f}  predicted house price = {:.2f}'.format(X3[44][0], predicted_charge[0]))

```

```python
#testing test dataset
X_test= test_df['X2 house age']
y_test = test_df['Y price sq feet']
X_test= np.array(X_test).reshape(-1,1) # reshape(-1,1) makes the array horizontal, y is vertical
y_pred = model_X2.predict(X_test)
r_tsq2 = model_X2.score(X_test, y_test)
print(r_tsq2)
plt.scatter(X_test, y_test,color='black', label='observed')
plt.plot(X_test, y_pred, label='predicted', color='Green', linewidth=2)
plt.xlabel('X2 house age')
plt.ylabel('Y price sq feet')
plt.title('Regression on testing data')
plt.legend(loc='best')
plt.show()
```
